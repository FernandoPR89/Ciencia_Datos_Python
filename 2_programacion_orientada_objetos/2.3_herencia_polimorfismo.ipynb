# Herencia

La herencia es otro concepto importante en POO, que nos permite crear nuevas clases, basadas en otras clases existentes.

La nueva clase hereda los atributos y métodos de la clase base y puede agregar sus propios atributos y métodos. De esta forma, la subclase puede “extender” o modificar el comportamiento de la superclase.

- Superclase: La clase de la cual se heredan los métodos y atributos.
- Subclase: La clase que hereda métodos y atributos de la superclase

En Python, para hacer que una clase herede de otra añadimos el nombre de la super clase entre paréntesis justo tras la definición de la clase. Así

```python
# Sintaxis
class subclase(superclase):
   # cuerpo de subclase
```

```python
# Definición de la clase Persona
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
    
    def mostrar_informacion(self):
        print(f"Nombre: {self.nombre}, Edad: {self.edad}")
```

```python
# Definición de la clase Estudiante que hereda de Persona
class Estudiante(Persona):
    def __init__(self, nombre, edad, nota, matricula):
        super().__init__(nombre, edad)
        self.nota = nota
        self.matricula = matricula

# Crear una instancia de Estudiante
estudiante = Estudiante("Luis", 20, "8.7", "MAT001")
estudiante.mostrar_informacion()
print(f'Nota: {estudiante.nota}, Matrícula: {estudiante.matricula}')
```

La función `super()` se utiliza para llamar al constructor de la clase base Persona, lo que permite inicializar nombre y edad.

## Sobreescritura de métodos

```python
class Estudiante(Persona):
    def __init__(self, nombre, edad, nota):
        super().__init__(nombre, edad)
        self.nota = nota
    
    # Sobreescribir el método mostrar_informacion
    def mostrar_informacion(self):
        print(f"Nombre: {self.nombre}, Edad: {self.edad}, Nota: {self.nota}")

# Crear una instancia de Estudiante
estudiante = Estudiante("Luis", 20, "8.7")
estudiante.mostrar_informacion()   # imprime 'Nombre: Luis, Edad: 20, Nota: 8.7'
```

## Llamar a métodos de la super clase

La clase derivada puede llamar a métodos de su clase base usando el método `super()`. De hecho, es habitual hacerlo con el método constructor __init__ para inicializar los atributos de la clase base.

Pero cualquier método de la clase base puede ser invocado

```python
# Definición de la clase Estudiante que hereda de Persona
class Estudiante(Persona):
    def __init__(self, nombre, edad, nota):
        super().__init__(nombre, edad)  # Llamada al constructor de la clase base
        self.nota = nota
    
    def mostrar_informacion(self):
        super().mostrar_informacion()  # Llamada a un método de la clase base
        print(f"Nombre: {self.nombre}, Edad: {self.edad}, Nota: {self.nota}")

# Crear una instancia de Estudiante
estudiante = Estudiante("Luis", 20, "8.7")
estudiante.mostrar_informacion()
```

# Polimorfismo

El polimorfismo es la capacidad de un objeto para tomar diferentes formas según el contexto.

El funcionamiento de un objeto se determina por su comportamiento, en lugar de su tipo específico.

```python
# Definición de la clase Persona
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
    
    def saludar(self):
        print(f"Nombre: {self.nombre}, Edad: {self.edad}")
```

```python
# Definición de la clase Profesor que hereda de Persona
class Profesor(Persona):
    def __init__(self, nombre, edad):
        super().__init__(nombre, edad)        
    
    def saludar(self):
        print("Soy un profesor")
        
# Definición de la clase Estudiante que hereda de Persona
class Estudiante(Persona):
    def __init__(self, nombre, edad):
        super().__init__(nombre, edad)        
    
    def saludar(self):
        print("Soy un estudiante")

class Vehiculo:
    def conducir(self):
        print("Conduciendo un vehículo")

# Función que usa polimorfismo
def imprimir_informacion(instancia):
    instancia.saludar()

# Crear instancias de Persona y Estudiante
persona = Persona("Carlos", 40)
profesor = Profesor("María", 30)
estudiante = Estudiante("Luis", 20)

auto = Vehiculo()

# Usar polimorfismo
imprimir_informacion(persona)
imprimir_informacion(profesor)
imprimir_informacion(estudiante)
imprimir_informacion(auto)
```
