## Creación y definición de strings

Un string es una secuencia de caracteres (letras, números, símbolos). Se define usando comillas simples 'texto', dobles "texto", o triples '''texto''' para textos multilínea.

```python
mensaje = "Hola, Python"
parrafo = '''Este es
un texto
multilínea'''
print(mensaje)
print(parrafo)
```

## Concatenación y repetición

- Concatenación: unir dos o más strings con el operador +.
- Repetición: duplicar un string varias veces con el operador *.

```python
saludo = "Hola" + " " + "Mundo"
eco = "Python " * 3
print(saludo)
print(eco)
```

## Indexación

La indexación permite acceder a un carácter específico dentro de un string usando su posición (índice).

```python
texto = "Python"
print(texto[0])   # 'P'
print(texto[-1])  # 'n'
```

## Funciones integradas
Son funciones integradas en Python que facilitan trabajar con strings.

- `.upper()`→ Convierte todo el texto a mayúsculas.
- `.lower()`→ Convierte todo el texto a minúsculas.
- `.capitalize()`→ Convierte la primera letra en mayúscula y el resto en minúsculas.
- `.strip()`→ Elimina espacios en blanco al inicio y al final.
- `.replace()`→ Reemplaza una parte del texto por otra.
- `.split()`→ Divide un string en una lista, usando un separador.
- `.join()`→ Une los elementos de una lista en un solo string, usando un separador.

```python
print("hola".upper())
print("HOLA".lower())
print("python".capitalize())
print("  hola  ".strip())
print("hola mundo".replace("mundo", "Python"))
print("uno,dos,tres".split(","))
print("-".join(["uno", "dos"]))

mi_cadena = "hola".upper()
mi_cadena = "HOLA".lower()
mi_cadena = "python".capitalize()
mi_cadena = "  hola  ".strip()
mi_cadena = "hola mundo".replace("mundo", "Python")
mi_cadena = "uno,dos,tres".split(",") 
mi_cadena = "-".join(["uno", "dos"])
```

## Interpolación y formato de strings
Es la forma de insertar valores dentro de un string.
La más usada es con f-strings:

```python
nombre = "Ana"
edad = 25
print(f"Me llamo {nombre} y tengo {edad} años")
```

## Comprobaciones

- `.startswith("Py")` → verifica si el texto empieza con "Py".
- `.endswith("on")` → verifica si el texto termina con "on".
- `.isdigit()` → verifica si el texto contiene solo números.
- `.isalpha()` → verifica si el texto contiene solo letras.

```python
"Python".startswith("Py")
"123".isdigit()
```
