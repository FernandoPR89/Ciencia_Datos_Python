# Procesamiento de listas



## Matrices
Una matriz es una estructura de datos que organiza elementos en forma de tabla o arreglo.
En Python no existe un tipo de dato “matriz” como tal, pero se pueden representar usando listas (listas anidadas) o librerías como NumPy.

### Matrices unidimensionales
Son listas simples, con una sola dimensión (una fila de elementos)

```python
# Matriz unidimensional (lista simple)
numeros = [1, 2, 3, 4, 5]
print(numeros[2])
```

### Matrices bidimensionales
Son listas dentro de listas, que representan filas y columnas (como una tabla)

```python
# Matriz bidimensional (lista de listas)
matriz = [
    [1, 2, 3],   # fila 0
    [4, 5, 6],   # fila 1
    [7, 8, 9]    # fila 2
]

print(matriz[0][1])
print(matriz[2][0])
```

### Matrices tridimensionales
Son listas de listas de listas, que representan datos en tres dimensiones (como un cubo).

```python
# Matriz tridimensional (lista de listas de listas)
cubo = [
    [ [1,2], [3,4] ],   # capa 0
    [ [5,6], [7,8] ]    # capa 1
]

print(cubo[0][1][0])
print(cubo[1][0][1])
```

## Slicing avanzado
Sublistas con saltos: 
- lista[::2] (pares)
- lista[::-1] (invertir)

```python
numeros = [0,1,2,3,4,5]
print(numeros[::2])   # [0,2,4]
print(numeros[::-1])  # [5,4,3,2,1,0]
```

## Listas por comprensión

Forma compacta de crear listas.

```python
cuadrados = [x**2 for x in range(5)]
print(cuadrados)
```

Por condiciones

```python
pares = [x for x in range(10) if x % 2 == 0]
```

## Funciones built-in aplicadas a listas

- `len(lista)` → longitud
- `sum(lista)` → suma de elementos
- `max(lista)` / min(lista) → máximo y mínimo
- `sorted(lista)` → lista ordenada sin modificar la original

```python
lista_numeros = [1, 2, 3, 4, 5]
print(len(lista_numeros))
print(sum(lista_numeros))
print(max(lista_numeros))
print(min(lista_numeros))
print(sorted(lista_numeros, reverse=True))
```
