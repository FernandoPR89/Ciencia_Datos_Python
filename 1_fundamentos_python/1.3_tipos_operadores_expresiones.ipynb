# 1.3 Tipos, operadores y expresiones

**Objetivo**:  Dominar tipos de datos y operaciones fundamentales

| **Tipos de datos**        | **Operadores**                                   |
| ------------------------- | ------------------------------------------------ |
| `int`, `float`, `complex` | Aritméticos: `+`, `-`, `*`, `/`, `//`, `%`, `**` |
| `str` (cadenas)           | Comparación: `==`, `!=`, `<`, `>`                |
| `bool`                    | Lógicos: `and`, `or`, `not`                      |
| `None`                    | Asignación: `=`, `+=`, `*=`                      |
