## Tipos de datos

**int**

Representa números sin decimales.

```python
edad = 25
print(edad) 
```

**float**

Representa números con punto decimal.

```python
pi = 3.14
print(pi)
```

**complex**

Representa números con parte real e imaginaria.

```python
z = 2 + 3j
print(z)        # (2+3j)
print(z.real)   # 2.0
print(z.imag)   # 3.0
```

**str** (cadena de texto)

Representa texto entre comillas simples o dobles.

```python
nombre = "Fernando"
direccion = 'Calle Palmira 123'
telefono = "5551234"
```

**bool** (valores lógicos)

Solo puede ser `True` o `False`

```python
es_mayor = True
print(es_mayor)
```

**None** (asusencia de valor)

Representa "nada" o valor vacio.

```python
resultado = None
print(resultado)
```

## Tipos de operadores

**Aritméticos**  
Se usan para hacer cálculos matemáticos.

| Operador | Descripción | Ejemplo |
|----------|-------------|---------|
| `+` | Suma | `2 + 3  # 5` |
| `-` | Resta | `7 - 4  # 3` |
| `*` | Multiplicación | `3 * 4  # 12` |
| `/` | División | `10 / 2  # 5.0` |
| `//` | División entera | `10 // 3  # 3` |
| `%` | Módulo (resto) | `10 % 3  # 1` |
| `**` | Potencia | `2 ** 3  # 8` |

```python
#Ejemplos de operadores aritmeticos
a = 10
b = 5
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a % b)
print(a ** b)
```

**De comparación**  
Sirven para comparar valores y devuelven `True` o `False`.

| Operador | Descripción | Ejemplo |
|----------|-------------|---------|
| `==` | Igualdad | `5 == 5  # True` |
| `!=` | Diferente | `5 != 3  # True` |
| `>` | Mayor que | `7 > 4  # True` |
| `<` | Menor que | `2 < 8  # True` |
| `>=` | Mayor o igual | `5 >= 5  # True` |
| `<=` | Menor o igual | `3 <= 4  # True` |

```python
# Ejemplo de operadores de comparación
x = 10
y = 5
print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

**Lógicos**  
Se usan para combinar condiciones.

| Operador | Descripción | Ejemplo |
|----------|-------------|---------|
| `and` | Verdadero si ambas condiciones lo son | `(5 > 3) and (2 < 4)  # True` |
| `or` | Verdadero si al menos una condición lo es | `(5 > 3) or (2 > 10)  # True` |
| `not` | Niega el valor lógico | `not (5 > 3)  # False` |

```python
# Ejemplo de operadores lógicos
p = True
q = False
print(p and q)
print(p or q)  
print(not p)   
```

**De asignación**  
Sirven para dar valores a variables y también para actualizar esos valores.

| Operador | Descripción | Ejemplo |
|----------|-------------|---------|
| `=` | Asignación simple | `x = 5` |
| `+=` | Suma y asigna | `x = 5; x += 2  # 7` |
| `-=` | Resta y asigna | `x = 5; x -= 2  # 3` |
| `*=` | Multiplica y asigna | `x = 3; x *= 4  # 12` |
| `/=` | Divide y asigna | `x = 10; x /= 2  # 5.0` |
| `//=` | División entera y asigna | `x = 10; x //= 3  # 3` |
| `%=` | Módulo y asigna | `x = 10; x %= 3  # 1` |
| `**=` | Potencia y asigna | `x = 2; x **= 3  # 8` |

```python
# Ejemplo de oeperadores de asignación
c = 10
c += 5
print("Suma y asignación:", c)
c -= 3
print("Resta y asignación:", c)
c *= 2
print("Multiplicación y asignación:", c)
c /= 4
print("División y asignación:", c)
c %= 3
print("Módulo y asignación:", c)
c **= 2
print("Potencia y asignación:", c)
```

En resumen:  
- **Aritméticos** → hacen cuentas.  
- **Comparación** → responden con verdadero o falso.  
- **Lógicos** → combinan condiciones.  
- **Asignación** → guardan o actualizan valores en variables.  

## Conversión de tipos de datos

La conversión de tipos de datos en Python significa transformar un valor de un tipo a otro usando funciones especiales. Esto se conoce como type casting. Es útil cuando necesitas que un dato se comporte como otro tipo (por ejemplo, convertir texto en número para hacer cálculos).

| Función | Descripción | Ejemplo | Resultado |
|---------|-------------|---------|-----------|
| `int()` | Convierte a número entero | `int("10")` | `10` |
| `float()` | Convierte a número decimal | `float("3.14")` | `3.14` |
| `str()` | Convierte a texto | `str(25)` | `"25"` |
| `bool()` | Convierte a valor lógico (True/False) | `bool(0)` | `False` |
|         |             | `bool("Hola")` | `True` |


```python
print(bool(0))       # False
print(bool(5))       # True
print(bool(""))      # False
print(bool("Hola"))  # True
```

## Métodos de tratamiento de cadenas

| Método / Sintaxis | Descripción | Ejemplo | Resultado |
|-------------------|-------------|---------|-----------|
| `.upper()` | Convierte todo el texto a mayúsculas | `"hola".upper()` | `"HOLA"` |
| `.lower()` | Convierte todo el texto en minúsculas | `"HOLA".lower()` | `"hola"` |
| `.capitalize()` | Convierte la letra inicial en mayúscula | `"hola".capitalize()` | `"Hola"` |
| `.replace()` | Reemplaza partes de la cadena por otra | `"Hola mundo".replace("mundo", "Python")` | `"Hola Python"` |
| `.split()` | Divide la cadena en una lista, usando un separador (por defecto espacio) | `"uno dos tres".split()` | `["uno", "dos", "tres"]` |
| `.strip()` | Remueve espacios en blanco al inicio y al final de un texto | `"   hola   ".lower()` | `"hola"` |

```python
#Ejemplo de metodos de tratamiento de cadenas de texto
texto = "hola Mundo"
print(texto.upper())
print(texto.lower())
print(texto.capitalize())
print(texto.replace("Mundo", "Python"))
print(texto.split())
texto = "manzana,pera,naranja"
print(texto)
print(texto.split(','))
texto = "   hola Mundo      "
print(texto)
print(texto.strip())
```

El **slicing** en Python es muy versátil porque permite seleccionar partes de una cadena (o lista).

| Sintaxis | Descripción | Ejemplo | Resultado |
|----------|-------------|---------|-----------|
| `[inicio:fin]` | Selecciona desde `inicio` hasta `fin-1` | `"Programación"[0:2]` | `"Pr"` |
| `[:fin]` | Desde el inicio hasta `fin-1` | `"Programación"[:4]` | `"Prog"` |
| `[inicio:]` | Desde `inicio` hasta el final | `"Programación"[3:]` | `"gramación"` |
| `[::paso]` | Selecciona con intervalos | `"Programación"[::2]` | `"Pormcó"` |
| `[inicio:fin:paso]` | Combina rango e intervalos | `"Programación"[1:5:2]` | `"rg"` |
| `[-n:]` | Últimos `n` caracteres | `"Programación"[-3:]` | `"ión"` |
| `[:-n]` | Todo menos los últimos `n` | `"Programación"[:-3]` | `"Programac"` |
| `[::-1]` | Invierte la cadena | `"Programación"[::-1]` | `"nóicamargorP"` |

```python
# Ejemplo de slicing en cadenas de texto
lenguaje = "Cadena de texto"
print(lenguaje[0:2])
print(lenguaje[:4])
print(lenguaje[3:])
print(lenguaje[::2])
print(lenguaje[1:5:2])
print(lenguaje[-3:])
print(lenguaje[:-3])
print(lenguaje[::-1])
```
