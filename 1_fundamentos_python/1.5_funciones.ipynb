# 1.5 Funciones

**Objetivo**:  Escribir código modular y reutilizable.

**Contenido**:
- Definición: def nombre(parametro=valor_default):
- Argumentos: posicionales, nombrados, *args, **kwargs
- Ámbito de variables: local vs global (global)
- Funciones lambda: operaciones simples en una línea
- Funciones built-in : len(), range(), enumerate(), zip(), map(), filter()

## Definición

Las funciones son bloques de código que realizan una tarea específica y pueden ser reutilizados en diferentes partes de un programa.
Estructurar nuestro código en funciones nos ayuda a hacerlo más organizado, de forma que sea más modular, más sencillo de mantener, y más fácil de reutilizar.

## Funciones built-in

Funciones que están incorporadas en la biblioteca estándar de Python. Estás funciones se pueden utilizar sin necesidad de una declaración import.
Dentro de parentesis ( ) y separados por *comas* se pasan los argumentos (valores) requeridos para su funcionamiento.

| Función       | Descripción breve |
|---------------|------------------|
| `print()`     | Muestra datos en la consola. |
| `len()`       | Devuelve la longitud de un objeto (lista, cadena, etc.). |
| `type()`      | Retorna el tipo de un objeto. |
| `int()`       | Convierte un valor a entero. |
| `float()`     | Convierte un valor a número decimal. |
| `str()`       | Convierte un valor a cadena de texto. |
| `list()`      | Crea una lista a partir de un iterable. |
| `dict()`      | Crea un diccionario. |
| `set()`       | Crea un conjunto (colección sin duplicados). |
| `tuple()`     | Crea una tupla (colección inmutable). |
| `range()`     | Genera una secuencia de números enteros. |
| `sum()`       | Devuelve la suma de los elementos de un iterable. |
| `max()`       | Retorna el valor máximo de un iterable. |
| `min()`       | Retorna el valor mínimo de un iterable. |
| `abs()`       | Devuelve el valor absoluto de un número. |
| `round()`     | Redondea un número al entero más cercano o a decimales especificados. |
| `sorted()`    | Devuelve una lista ordenada a partir de un iterable. |
| `input()`     | Captura datos ingresados por el usuario. |
| `help()`      | Muestra la ayuda/documentación de objetos o funciones. |
| `isinstance()`| Verifica si un objeto es instancia de una clase o tipo. |


```python
pi = 3.14159265359
redondear = round(pi, 2) # Redondea el valor del primer argumento, a los decimales indicados en el segundo argumento
print('Valor redondeando de pi =', redondear)

lista = [1, 2, 3, 4, 5]
print('Devuelve la longitud de lista =', len(lista))
print('Devuelve el ID de lista =', id(lista))
print('Devuelve el tipo de dato de lista =', type(lista))
print('Devuelve el tipo de dato de lista =', type(pi))
print('Valor absoluto =', abs(-25))
print('Valor máximo =', max(lista))
print('Valor mínimo =', min(lista))

Num_Dec = 7
Num_Bin = bin(Num_Dec)
print('Valor binario de', Num_Dec, 'es', Num_Bin)

real = 2.34
imag = 4.56
print('Valor complejo =', complex(real, imag))
```

## Funciones definidas por el usuario
Bloques de código creados por el programador para necesidades específicas.
En Python, una función se define con la palabra clave **def**, seguida del nombre de la función y paréntesis que pueden contener los parámetros de entrada.

```python
def nombre_de_la_funcion(parametros):
    resultado = None  # Inicialización de una variable para almacenar el resultado
    # Bloque de código de la función
    # Puede contener múltiples instrucciones
    return resultado  # Devuelve un valor (opcional)
```

- **nombre_de_la_funcion**: Es el nombre que le damos a nuestra función.
- **parametros**: Son los parámetros que nuestra función puede recibir. Estos son opcionales y pueden ser utilizados para pasar información a la función.
- **return resultado**: La instrucción return nos permite devolver un resultado de la función. Es opcional y puede devolver cualquier tipo de dato.
