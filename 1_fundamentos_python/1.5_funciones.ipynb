## Definición

Las funciones son bloques de código que realizan una tarea específica y pueden ser reutilizados en diferentes partes de un programa.
Estructurar nuestro código en funciones nos ayuda a hacerlo más organizado, de forma que sea más modular, más sencillo de mantener, y más fácil de reutilizar.

## Funciones definidas por el usuario
Bloques de código creados por el programador para necesidades específicas.
En Python, una función se define con la palabra clave **def**, seguida del nombre de la función y paréntesis que pueden contener los parámetros de entrada.

````python
def nombre_de_la_funcion(parametros):
    resultado = None  # Inicialización de una variable para almacenar el resultado
    # Bloque de código de la función
    # Puede contener múltiples instrucciones
    return resultado  # Devuelve un valor (opcional)
````

- **nombre_de_la_funcion**: Es el nombre que le damos a nuestra función.
- **parametros**: Son los parámetros que nuestra función puede recibir. Estos son opcionales y pueden ser utilizados para pasar información a la función.
- **return resultado**: La instrucción return nos permite devolver un resultado de la función. Es opcional y puede devolver cualquier tipo de dato.

## Llamar a una función

Una vez que tenemos una función definida, esta puede ser llamada desde cualquier parte del programa (a esto también se le llama, invocar a la función).

Para llamar a una función, se utiliza su nombre seguido de paréntesis. En caso de que la función requiera argumentos, tendremos que ponerlos entre los paréntesis.

````python
def nombre_de_la_funcion():
    return "Hola"

nombre_de_la_funcion() # Llamada a la función
````

## Parámetros
Los parámetros son variables que podemos pasar a nuestra función al invocarla. Esto nos permite crear funciones flexibles y reutilizables.

````python
def nombre_de_la_funcion(parametro1, parametro2, parametro3):
    # bloque de código de la función que utiliza los parámetros
    pass
````

Por ejemplo, imaginemos que tenemos una función que simplemente multiplique dos números. Podemos hacer que reciba dos parámetros y realice la multiplicación.

````python
def multiplicacion(a, b):
    print(a * b) # Output: 15

multiplicacion(3, 5) # llamamos a la función
````

### Parámetros posicionales

Los parámetros posicionales son los parámetros más comunes. Se definen en el orden en que aparecen en la lista de parámetros de la función y se pasan a la función en el mismo orden.

````python
def operacion(a, b, c):
    return a * b - c

# Llamada a la función con parámetros posicionales
resultado_multiplicacion = operacion(5, 3, 2)
print(resultado_multiplicacion)  # Salida: 13
resultado_multiplicacion = operacion(2, 3, 5)
print(resultado_multiplicacion)  # Salida: 1
````

### Parametros de longitud variable
Python permite definir funciones que aceptan un número variable de parámetros utilizando *args y **kwargs.

*args: Se utiliza para pasar una lista de argumentos posicionales

````python
def suma_numeros(*args):
    resultado = sum(args)
    return resultado

# Llamada a la función con una cantidad variable de argumentos
resultado_suma = suma_numeros(1, 2, 3, 4, 5)
print(resultado_suma)  # Salida: 15
````

**kwargs: Se utiliza para pasar un diccionario de argumentos de palabra clave (keywords arguments)

````python
def imprimir_info(**kwargs):
    for clave, valor in kwargs.items():
        print(clave + ":", valor)

# Llamada a la función con argumentos con nombre
imprimir_info(nombre="Fernando", edad=36, ciudad="Cuernavaca")
imprimir_info(ciudad="Cuernavaca", edad=36, nombre="Fernando")
````

Pasar la instancia de un objeto como parámetro

````python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

persona1 = Persona("Fernando", 36)

def saludar(persona):
    return f"Hola, {persona.nombre}!"

print(saludar(persona1))  # Salida: "Hola, Fernando!"
````

### Paso por valor o por referencia
 En Python, el paso de argumentos a funciones se realiza siempre por valor (es decir, que la función siempre recibe una copia del argumento).

Pero el comportamiento será diferente si lo que pasamos es un tipo básico o una referencia.

En un **tipo básico**, la función invocada no podrá modificar el argumento que recibe
En un **tipo referencia**, sí podrá modificar el argumento que recibe.

````python
# Ejemplo con un tipo básico (int)
def modificar_numero(numero):
    numero = numero * 2

numero_original = 5
modificar_numero(numero_original)
print("Número original después de llamar a la función:", numero_original)
````

````python
# Ejemplo con una referencia (lista)
def modificar_lista(lista):
    lista.append(4)

lista_original = [1, 2, 3]
modificar_lista(lista_original)
print("Lista original después de llamar a la función:", lista_original)
````
