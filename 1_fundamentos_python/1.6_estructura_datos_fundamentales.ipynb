# 1.6 Estructura de control de flujo fundamentales

**Objetivo**: Controlar el flujo de ejecución del programa.

**Contenido**:
- Condicionales: `if`, `elif`, `else`, operador ternario
- Bucles: `for` (iteraciones definidas), `while` (condicionales)
- Control de bucles: `break`, `continue`, `pass`, else en bucles
- Iteradores: `range()`, `enumerate()`, `zip()`

## Condicionales
Los condicionales son estructuras de control que permiten que un programa tome decisiones basadas en ciertas condiciones.

Estas sentencias nos permiten evaluar expresiones y ejecutar diferentes bloques de código dependiendo de si estas expresiones son verdaderas o falsas.

En Python, los condicionales se implementan principalmente a través de las sentencias `if`, `elif`, y `else`

### Sentencia if
La sentencia if se utiliza para ejecutar un bloque de código si una condición es verdadera. La sintaxis básica es la siguiente:

```python
# Sintaxis
if condicion:
    # bloque de código si la condición es verdadera
```

```python
numero = 15

if numero > 10:
    print("El número es mayor que 10")
```

### Sentencia else
La sentencia `else` se utiliza para ejecutar un bloque de código cuando la condición en el `if` es falsa. La sintaxis es la siguiente:

```python
# Sintaxis
if condicion:
    # bloque de código si la condición es verdadera
else:
    # bloque de código si la condición es falsa
```

```python
numero = 5

if numero > 10:
    print("El número es mayor que 10")
else:
    print("El número es menor o igual a 10")
```

### Sentencia elif
La sentencia elif se utiliza para evaluar múltiples condiciones de manera secuencial.

```python
# Sintaxis
if condicion_1:
    # bloque de código si la condicion_1 es verdadera
elif condicion_2:
    # bloque de código si la condicion_1 es falsa y la condicion_2 es verdadera
else:
    # bloque de código si todas las condiciones anteriores son falsas
```

```python
numero = 5

if numero > 0:
    print("El número es positivo")
elif numero < 0:
    print("El número es negativo")
else:
    print("El número es cero")
```

```python
lista = [1, 2, 3, 4]
for elemento in lista:
    if elemento == 1:
        print(elemento * 2)
    elif elemento == 2:
        print(elemento * 2) 
    elif elemento == 3:
        print(elemento * 3)
    elif elemento == 4:
        print(elemento * 2)
    else:
        print("Elemento no reconocido")

for elemento in lista:
    if elemento==3:
        print(elemento * 3)
    print(elemento * 2)
```

### Condiciones múltiples
También es posible combinar múltiples condiciones utilizando los operadores `and`, `or` y `not`. Estos operadores nos permiten construir condiciones más complejas:

`and` : Retorna `True` si ambas condiciones son verdaderas.
`or` : Retorna `True` si al menos una de las condiciones es verdadera.
`not` : Retorna `True` si la condición es falsa.

```python
x = 10
y = 5
z = 0

if x > y and y > z:
    print("Todas las condiciones son verdaderas")
if x > y or y < z:
    print("Al menos una de las condiciones es verdadera")
    
if not x > y:
    print("El número x no es mayor que y")
else:
    print("El número x es mayor que y")
print(not x > y)
```

## Bucles
Los bucles son una herramienta que nos permite repetir un bloque de código varias veces. En Python, contamos con dos tipos principales de bucles: `for` y `while`.

### Bucle for
El bucle `for` se utiliza para iterar sobre una secuencia (como una lista, tupla, diccionario, etc.) y ejecutar un bloque de código para cada elemento en esa secuencia.

La sintaxis básica del bucle `for` es la siguiente:

```python
# Sintaxis
for elemento in secuencia:
    # bloque de código a ejecutar para cada elemento
```

```python
# Iterar sobre una colección de tipo lista
frutas = ["manzana", "banana", "cereza"]

for fruta in frutas:
    print(fruta)
```

```python
# Iterar un número de veces
for i in range(5):
    print(i)
```

```python
# Iterar un sobre una cadena de texto
for letra in "Python":
    print(letra)
```

### Bucle while
El bucle `while` se utiliza para repetir un bloque de código mientras una condición sea verdadera. La sintaxis básica del bucle `while` es la siguiente:

```python
# Sintaxis
while condicion:
    # bloque de código a ejecutar mientras la condición sea verdadera
```

```python
# Ejemplo de bucle while para contar de 1 a 10
contador = 1

while contador <= 10:
    print(contador) 
    contador += 1
```

```python
# Ejemplo de bucle while para iterar sobre una lista
frutas = ["manzana", "banana", "cereza"]
frutas.append("naranja")  # Agregar un elemento a la lista
while frutas:
    fruta = frutas.pop() # El método pop() elimina y devuelve el último elemento de la lista
    print(fruta)
```

## Control de bucles break, continue y pass
Dentro de los bucles en Python, también podemos utilizar las sentencias `break` y `continue` para controlar el flujo de ejecución.

**break** Termina el bucle y ejecuta el bloque de código que está después del bucle.

```python
# El bucle for se detiene cuando encuentra la fruta "sandía"
frutas = ["manzana", "banana", "cereza", "sandía", "uva"]

for fruta in frutas:
    print(fruta)
    if fruta == "sandía":
        break
```

**continue** nos permite saltar a la siguiente iteración sin ejecutar el resto del bloque de código para esa iteración.

```python
# for solo imprime los números impares de la lista
numeros = [1, 2, 3, 4, 5]

for numero in numeros:
    if numero % 2 == 0:
        continue
    print(numero)
```

**pass** es la sentencia nula de Python: no hace absolutamente nada. Su única función es satisfacer el requisito sintáctico de que cualquier bloque abierto con dos puntos (:) contenga al menos una sentencia. Sin pass, un bloque vacío produce SyntaxError.

```python
# Funcion que no hace nada, pero es un marcador de posición para código que se implementará en el futuro
def funcion_pendiente():
    pass    # implementar en la próxima sesión
```

### Bucles con else
En Python, los bucles también pueden tener una cláusula `else`, que se ejecuta cuando el bucle termina sin haber sido interrumpido por un `break`

```python
# El bucle for-else se ejecuta normalmente, pero si se encuentra un break, el bloque else no se ejecuta
nombres = ["Alice", "Bob", "Charlie", "David", "Eve"]
bandera = False
for nombre in nombres:
    if nombre == "David":
        break
else:
    bandera = True

if bandera:
    print("El bucle ha terminado sin encontrar el nombre 'German'")

if 'German' not in nombres:
    print("El bucle ha terminado sin encontrar el nombre 'German'")
```

## Iteradores range(), enumerate(), zip()


### Range ( )

La función `range()` en Python genera una secuencia inmutable de números enteros en un rango específico.

```python
#sintaxis
range(inicio, fin, paso)
```

- **inicio**: Valor inicial de la secuencia. Si no se especifica, por defecto es 0.
- **fin**: Valor final de la secuencia. Este valor nunca es parte de la secuencia generada.
- **paso**: Incremento entre cada número de la secuencia. Si no se especifica, por defecto es 1.

```python
# range(fin): de 0 a 4
for i in range(5): # va de 0 a 4, el valor final no se incluye
    print(i)          # 0 1 2 3 4
print("\n")

# range(inicio, fin): el valor final no se incluye, el bucle va de inicio a fin-1
for i in range(1, 6):
    print(i)          # 1 2 3 4 5
print("\n")

# range(inicio, fin, paso): con incremento
for i in range(0, 20, 5):
    print(i)          # 0 5 10 15
print("\n")

# Contar hacia atrás (paso negativo)
for i in range(10, 0, -1):
    print(i)          # 10 9 8 7 ... 1
print("\n")
```

### enumerate ( )

`enumerate()` devuelve pares (índice, valor) para cada elemento. Es la forma idiomática de Python cuando necesitas el índice y el valor simultáneamente — evita el patrón range(len(...)) que es más verboso y propenso a errores.

```python
# Sintaxis
enumerate(iterable, start=0)
```

```python
frutas = ["manzana", "pera", "naranja", "uva"]

# Forma no idiomática (viene de C/Java):
for i in range(len(frutas)):
    print(i, frutas[i])

print("\n")

# Forma idiomática Python:
for i, fruta in enumerate(frutas):
    print(i, fruta)

print("\n")

# Empezar desde 1 en lugar de 0:
for i, fruta in enumerate(frutas, start=1):
    print(f"{i}. {fruta}")
```

### zip ( )
`zip` combina dos (o más) secuencias en pares, creando tuplas con los elementos correspondientes.

```python
# Sintaxis
zip(iterable1, iterable2, ...)
```

```python
# Iterar sobre dos listas al mismo tiempo
nombres = ["Ana", "Luis", "Pedro"]
edades = [20, 25, 30]
colroes = ["rojo", "verde", "azul"]

for nombre, edad in zip(nombres, edades):
    print(nombre, edad)

print("\n")

for nombre, edad, color in zip(nombres, edades, colroes):
    print(nombre, edad, color)
```
