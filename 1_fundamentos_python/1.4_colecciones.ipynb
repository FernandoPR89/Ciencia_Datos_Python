# 1.4 Colecciones

**Objetivo**:  Manejar grupos de datos eficientemente.

- Listas []
- Tuplas ()
- Diccionarios {}
- Conjuntos set()

| **Estructura**        | **Características**             | **Cuándo usarla**                            |
| --------------------- | ------------------------------- | -------------------------------------------- |
| **Listas** `[]`       | Ordenada, mutable, duplicados   | Secuencias modificables                      |
| **Tuplas** `()`       | Ordenada, inmutable, duplicados | Datos constantes                             |
| **Diccionarios** `{}` | Pares clave-valor, mutable      | Búsquedas rápidas por clave                  |
| **Conjuntos** `set()` | No ordenado, únicos, mutable    | Eliminar duplicados, operaciones matemáticas |
