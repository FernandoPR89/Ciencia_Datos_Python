# Módulo 1: Entornos de Programación

**Objetivo**: Conocer y utilizar los conceptos básicos de Python en distintos entornos de programación.

**Entorno de programación** (*Integrated Development Environment*, IDE) es una aplicación de software que consolida herramientas esenciales para desarrolladores, como un editor de código, compilador/intérprete, depurador y,  menudo, control de versiones, en una sola interfaz para agilizar el desarrollo de software.
