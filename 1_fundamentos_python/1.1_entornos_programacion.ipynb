## Entorno de programación


El entorno de programación (del inglés, *Integrated Development Environment*, IDE) es una aplicación de software que consolida herramientas esenciales para desarrolladores tales como:
- **Editor de código**: Herramienta dentro del IDE que permite escribir y modificar programas, ofreciendo funciones como resaltado de sintaxis, autocompletado y formato del código.

```python
print("Bienvenido al curso de Fundamentos de Python para Ciencia de Datos")
```


- **Compilador/intérprete**: Componente que traduce el código fuente a un lenguaje que la computadora pueda ejecutar. El compilador genera un programa ejecutable, mientras que el intérprete lo ejecuta línea por línea

- **Depurador**: Módulo que ayuda a encontrar y corregir errores en el código, permitiendo ejecutar el programa paso a paso, inspeccionar variables y analizar el flujo de ejecución.

```python
x=1
y=3
z = x*3 + y*2
print(f" z = {z}")
```


- **Control de versiones**: Sistema que gestiona los cambios en el código a lo largo del tiempo, permitiendo guardar historial, colaborar en equipo y recuperar versiones anteriores (ejemplo: Git)

## Creación de entorno virtual

Un entorno virtual en Python es un espacio aislado donde puedes instalar paquetes y dependencias sin afectar el sistema global ni otros proyectos. Esto permite que cada proyecto tenga sus propias versiones de librerías, evitando conflictos.
Primero se crea:

`python -m venv nombre_del_entorno`

Luego se activa:

`nombre_del_entorno\Scripts\activate`

Por qué no se debe subir al repositorio
- El entorno virtual contiene archivos binarios y dependencias específicas del sistema operativo, lo que lo hace pesado e innecesario en el repositorio.
- Puede generar incompatibilidades entre diferentes sistemas.
- Lo correcto es subir solo el archivo **requirements.txt**, que lista las dependencias necesarias. Así, cada colaborador puede recrear el entorno en su propia máquina

## Creación de archivo de requirements.txt

El archivo requirements.txt en Python es un documento de texto que lista todas las dependencias (paquetes y librerías) necesarias para que un proyecto funcione. Sirve para que cualquier persona pueda recrear el mismo entorno de trabajo instalando exactamente las mismas versiones de los paquetes

Como se crea:

`pip freeze > requirements.txt`

Ejemplo de arhivo requerements.txt

`numpy==1.26.4`
`pandas==2.2.1`
`matplotlib==3.8.3`
`scikit-learn==1.4.0`

Como se utiliza en otro equipo:

`pip install -r requirements.txt`
